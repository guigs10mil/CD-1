### Variáveis escolhidas:

Taxa de mulheres acima de 15 anos empregadas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import *

%matplotlib inline
```

```python
data1 = pd.read_excel("indicator undata total_fertility.xlsx", index_col=0)
data2 = pd.read_excel("Years in school women 25 plus.xlsx", index_col=0)
data3 = pd.read_excel("indicator_f above 15 employ.xlsx", index_col=0)
```

```python
data1 = data1.loc[:,2007]
data2 = data2.loc[:,2007]
data3 = data3.loc[:,2007]
```

```python
data1 = data1.dropna()
data2 = data2.dropna()
data3 = data3.dropna()
```

```python
paises = []
var1 = []
var2 = []
var3 = []

for i in data1.keys():
    if i in data2.keys():
        if i in data3.keys():
            paises.append(i)
            var1.append(data1.get(i))
            var2.append(data2.get(i))
            var3.append(data3.get(i))

d = {"Var1": var1, "Var2": var2, "Var3": var3}
df = pd.DataFrame(data=d, index=paises)
df
```

```python
f, (ax1, ax2) = plt.subplots(1,2,figsize=(16,6))
ax1.scatter(df.Var2, df.Var1, alpha=0.7)
ax1.set_xlabel("Média de anos que as mulheres acima de 25 anos passaram na escola")
ax1.set_ylabel("Fertilidade (filhos por mulher)")

ax2.scatter(df.Var3, df.Var1, alpha=0.7)
ax2.set_xlabel("Taxa de mulheres acima de 15 anos empregadas (%)")
ax2.set_ylabel("Fertilidade (filhos por mulher)")


plt.tight_layout(1)
```
